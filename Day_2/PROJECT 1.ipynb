Write a function function for a 1D Position as a function of time and initial conditions
[constant acceleration]

```python
def x(x0,v0,t):
    return x0 + v0 * t
print(x(0,5,3))
```

```python
x0 = 0
v0 = 5
print(x(x0,v0,4))
```

Write a function function for a 1D Velocity as a function of time and initial conditions
[constant acceleration]

```python
def constant_acc(t,x0,v0,a):
    return x0 + v0 * t + 1/2 * a * t**2
print(constant_acc(3,0,10,5))
```

```python
x0 = 0
a = 5
print(constant_acc(5,x0,20,a))
```

```python
import numpy as np
t = np.arange(0,0.7,0.01)
print(t)
```

```python
positions = constant_acc(t,1,10,5)
print(constant_acc(t,1,10,5))
```

```python
import matplotlib.pyplot as plt
plt.plot(t, positions)
plt.show()
```

Plotting some projectile motion:

```python
projectile = constant_acc(t,1.8,0,-9.8)
plt.plot(t, projectile)
plt.xlabel('Time')
plt.ylabel('Distance travelled')
plt.title('Projectile Motion')
plt.show()
```
